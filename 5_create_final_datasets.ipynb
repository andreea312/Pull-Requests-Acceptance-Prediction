```python
import pandas as pd
```

```python
AUGMENTED_CSV = f"final_datasets/test/augmented.csv"
FINAL_CSV = f"final_datasets/test/final.csv"
```

```python
df = pd.read_csv(AUGMENTED_CSV)
df
```

```python
df = df.drop(columns=[c for c in df.columns.tolist() if c not in ['author_acceptance_rate', 'core_contributor_flag', 'previous_prs', 'year_created', 'time_to_close', 'comments_burstiness', 'num_comments', 'num_reviewers', 'avg_max_args', 'avg_multi_comments', 'max_max_cc', 'min_call_count', 'pr_time_label']])
```

```python
df.replace(-1, pd.NA, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Rows: {len(df)}  Columns: {len(df.columns)}")
```

```python
df['pr_time_label'] = df['pr_time_label'].replace({'immediate': 'accepted', 'fast': 'accepted', 'moderate': 'accepted', 'long': 'accepted', 'never': 'rejected'})
print(df['pr_time_label'].unique())
df['pr_time_label'].value_counts()
```
