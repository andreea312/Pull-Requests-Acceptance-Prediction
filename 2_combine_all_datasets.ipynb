```python
SCOPE = "train_val"
csv_files = [f"{COMBINED_FOLDER}/{SCOPE}/django_django_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/keras-team_keras_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/fastai_fastai_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/matplotlib_matplotlib_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/scrapy_scrapy_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/sqlalchemy_sqlalchemy_combined.csv",
             f"{COMBINED_FOLDER}/{SCOPE}/tiangolo_fastapi_combined.csv"]
```
